```python
# Task 1: Create the dataset using pandas


import pandas as pd

data = {
    "cgpa":         [6.5, 7.8, 5.2, 8.9, 6.0, 9.1, 7.2, 5.8, 8.5, 7.0],
    "iq":           [110, 125, 95, 140, 105, 145, 120, 100, 135, 115],
    "grade":        ["B", "A", "C", "A", "B", "A", "B", "C", "A", "B"],
    "placement":    ["yes", "yes", "no", "yes", "no", "yes", "yes", "no", "yes", "yes"]
}

df = pd.DataFrame(data)


# Task 2: Apply Ordinal Encoding to 'grade' (C < B < A)

grade_mapping = {"C": 0, "B": 1, "A": 2}
df["grade"] = df["grade"].map(grade_mapping)


# Task 3: Apply Label Encoding to 'placement'

placement_mapping = {"no": 0, "yes": 1}
df["placement"] = df["placement"].map(placement_mapping)


# Task 4: Separate features (X) and target (y)
# X contains input variables, y contains output variable

X = df[["cgpa", "iq", "grade"]]
y = df["placement"]


# Task 5: Split data into training and testing sets (90-10 split)
# Training data is used to train the model, test data is used to evaluate it

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42
)


# Task 6: Apply StandardScaler
# Scale features so they have mean=0 and variance=1 (important for Logistic Regression)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)  # fit only on training data
X_test_scaled = scaler.transform(X_test)        # transform test data


# Task 7: Train Logistic Regression model and evaluate accuracy
# Model learns from training data and predicts on test data

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)

print("Test Accuracy:", accuracy)
```
