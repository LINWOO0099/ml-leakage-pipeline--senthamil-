```python
import math

def sigmoid(z):
    """Compute sigmoid value for a given real number."""
    return 1 / (1 + math.exp(-z))


def predict_class(z):
    """Return class label (1 or 0) based on sigmoid probability."""
    prob = sigmoid(z)
    return 1 if prob >= 0.5 else 0


def evaluate(y_true, y_pred):
    """Calculate accuracy, false positives, and false negatives."""
    correct = 0
    false_positives = 0
    false_negatives = 0

    for actual, pred in zip(y_true, y_pred):
        if actual == pred:
            correct += 1
        elif actual == 0 and pred == 1:
            false_positives += 1
        elif actual == 1 and pred == 0:
            false_negatives += 1

    accuracy = correct / len(y_true)

    return {
        "accuracy": accuracy,
        "false_positives": false_positives,
        "false_negatives": false_negatives
    }


def display_report(y_true, y_pred, z_values):
    """Display probabilities, predictions, and evaluation metrics."""
    probabilities = [sigmoid(z) for z in z_values]
    predictions = [predict_class(z) for z in z_values]
    metrics = evaluate(y_true, predictions)

    print("Probabilities    :", probabilities)
    print("Predicted Classes:", predictions)
    print("Accuracy         : {:.2f}%".format(metrics["accuracy"] * 100))
    print("False Positives  :", metrics["false_positives"])
    print("False Negatives  :", metrics["false_negatives"])


# Test data
z_values = [2.5, -1.2, 0.0, 3.1, -2.8]
y_true   = [1, 0, 1, 1, 0]

# Run report
display_report(y_true, None, z_values)
```
